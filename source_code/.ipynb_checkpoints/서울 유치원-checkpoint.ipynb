```python
#def get_list(total_kinder_number):
kinder_id_list = []
list_url = 'http://e-childschoolinfo.moe.go.kr/kinderMt/kinderFind.do'
list_form = {
    'sidoList':11,
    'sggList':99,
    'establishmentType':99,
    'pageIndex':0
    } 

list_r = requests.get(list_url, data=list_form)		#유치원 별 url을 가져온다. 
while(True) :
    list_form['pageIndex'] += 1
    list_r = requests.get(list_url, data=list_form)
    soup = bs4.BeautifulSoup(list_r.content, 'html.parser')
    try :
        kinder_list = soup.find('ul', attrs={'class':'chkList'}).find_all('input')	#chk 항목을 가져온다.
    except :
        break
    for kinder in kinder_list:
        kinder_id_list.append(kinder.attrs['value'])	#모든 value를 가져온다.
    
kinder_id_list
```

```python
kinder_name = []
boss = []
director = []
established_type = []
established_date = []
start_date = []
admin_department = []
tel = []
homepage = []
operating_hour = []
address = []
service = []
age_3_class = []
age_3_student = []
age_4_class = []
age_4_student = []
age_5_class = []
age_5_student = []
age_3_4_class = []
age_3_4_student = []
age_4_5_class = []
age_4_5_student = []
age_3_5_class = []
age_3_5_student = []
handicap_class = []
handicap_student = []
staff_director = []
staff_sub_director = []
staff_chief_teacher = []
staff_manager_teacher = []
staff_teacher = []
staff_special_teacher = []
staff_health_teacher = []
staff_nutrient_teacher = []
staff_term_teacher = []
staff_office_worker = []
staff_chief_grade = []
staff_1st_grade = []
staff_2nd_grade = []
staff_basic_grade = []
under_1 = []
over_1_under_2 = []
over_2_under_4 = []
over_4_under_6 = []
over_6 = []
certification_date = []


from tqdm import tqdm_notebook

for kinder_id in tqdm_notebook(kinder_id_list):

    detail_url = 'http://e-childschoolinfo.moe.go.kr/kinderMt/kinderSummary.do'
    list_form = {
        'sidoList':99,
        'sggList':99,
        'establishmentType':99,
        'ittId':kinder_id,
        'pageIndex':1
        }
    detail_r = requests.post(detail_url, data=list_form)

    table = bs4.BeautifulSoup(detail_r.content, 'html.parser').find_all('tr')
    
    service_list = []
    if 'checked' in table[11].select('#preschool-service-1')[0].attrs:
        service_list.append('방과후과정')
    if 'checked' in table[11].select('#preschool-service-2')[0].attrs:
        service_list.append('특수학급')
    if 'checked' in table[11].select('#preschool-service-3')[0].attrs:
        service_list.append('아침돌봄')
    if 'checked' in table[11].select('#preschool-service-4')[0].attrs:
        service_list.append('저녁돌봄')
    if 'checked' in table[11].select('#preschool-service-5')[0].attrs:
        service_list.append('온종일돌봄')
    if 'checked' in table[11].select('#preschool-service-6')[0].attrs:
        service_list.append('해당없음')
        
        
    kinder_name.append(table[0].find('td').text.strip())
    boss.append(table[1].find('td').text.strip())
    director.append(table[2].find('td').text.strip())
    established_type.append(table[3].find('td').text.strip())
    established_date.append(table[4].find('td').text.strip())
    start_date.append(table[5].find('td').text.strip())
    admin_department.append(table[6].find('td').text.strip())
    tel.append(table[7].find('td').text.strip())
    homepage.append(table[8].find('td').text.strip())
    operating_hour.append(table[9].find('td').text.strip())
    address.append(table[10].find('td').text.strip())
    service.append(service_list)
    age_3_class.append(table[14].find_all('td')[0].text.strip().replace(',',''))
    age_4_class.append(table[14].find_all('td')[1].text.strip().replace(',',''))
    age_5_class.append(table[14].find_all('td')[2].text.strip().replace(',',''))
    age_3_4_class.append(table[14].find_all('td')[3].text.strip().replace(',',''))
    age_4_5_class.append(table[14].find_all('td')[4].text.strip().replace(',',''))
    age_3_5_class.append(table[14].find_all('td')[5].text.strip().replace(',',''))
    handicap_class.append(table[14].find_all('td')[6].text.strip().replace(',',''))

    age_3_student.append(table[15].find_all('td')[0].text.strip().replace(',',''))
    age_4_student.append(table[15].find_all('td')[1].text.strip().replace(',',''))
    age_5_student.append(table[15].find_all('td')[2].text.strip().replace(',',''))
    age_3_4_student.append(table[15].find_all('td')[3].text.strip().replace(',',''))
    age_4_5_student.append(table[15].find_all('td')[4].text.strip().replace(',',''))
    age_3_5_student.append(table[15].find_all('td')[5].text.strip().replace(',',''))
    handicap_student.append(table[15].find_all('td')[6].text.strip().replace(',',''))

    certification_date.append(table[18].find_all('td')[1].text.strip().replace(',',''))

    staff_url = 'http://e-childschoolinfo.moe.go.kr/kinderMt/kinderChildAndStaff.do'
    list_form = {
        'ittId':kinder_id,
        'pageIndex':1
        }
    staff_r = requests.post(staff_url, data=list_form)

    staff = BeautifulSoup(staff_r.content, 'html.parser')

    staff_director.append(staff.select('.content')[1].select('td')[1].text.strip().replace(',',''))
    staff_sub_director.append(staff.select('.content')[1].select('td')[2].text.strip().replace(',',''))
    staff_chief_teacher.append(staff.select('.content')[1].select('td')[3].text.strip().replace(',',''))
    staff_manager_teacher.append(staff.select('.content')[1].select('td')[4].text.strip().replace(',',''))
    staff_teacher.append(staff.select('.content')[1].select('td')[5].text.strip().replace(',',''))
    staff_special_teacher.append(staff.select('.content')[1].select('td')[6].text.strip().replace(',',''))
    staff_health_teacher.append(staff.select('.content')[1].select('td')[7].text.strip().replace(',',''))
    staff_nutrient_teacher.append(staff.select('.content')[1].select('td')[8].text.strip().replace(',',''))
    staff_term_teacher.append(staff.select('.content')[1].select('td')[9].text.strip().replace(',',''))
    staff_office_worker.append(staff.select('.content')[1].select('td')[10].text.strip().replace(',',''))

    staff_chief_grade.append(staff.select('.content')[1].select('td')[11].text.strip().replace(',',''))
    staff_1st_grade.append(staff.select('.content')[1].select('td')[12].text.strip().replace(',',''))
    staff_2nd_grade.append(staff.select('.content')[1].select('td')[13].text.strip().replace(',',''))
    staff_basic_grade.append(staff.select('.content')[1].select('td')[13].text.strip().replace(',',''))

    under_1.append(staff.select('.content')[2].select('td')[0].text.strip().replace(',',''))
    over_1_under_2.append(staff.select('.content')[2].select('td')[1].text.strip().replace(',',''))
    over_2_under_4.append(staff.select('.content')[2].select('td')[2].text.strip().replace(',',''))
    over_4_under_6.append(staff.select('.content')[2].select('td')[3].text.strip().replace(',',''))
    over_6.append(staff.select('.content')[2].select('td')[4].text.strip().replace(',',''))
```

```python
from tqdm import tqdm_notebook

lat = []
lng = []

for n in tqdm_notebook(kinder_info.index):
    try:
        tmp_add = str(kinder_info['주소'][n]).split('(')[0]
        tmp_map = gmaps.geocode(tmp_add)
        
        tmp_loc = tmp_map[0].get('geometry')
        lat.append(tmp_loc['location']['lat'])
        lng.append(tmp_loc['location']['lng'])
        
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        print("Here is nan !")
        
        
kinder_info['lat'] = lat
kinder_info['lng'] = lng
kinder_info.head()
```
